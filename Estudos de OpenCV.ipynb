# Experiência com OpenCV em python3

Certifique que as bibliotecas estajam instaladas:

```python
import cv2 as cv #Open CV
import numpy as np # Serve para cálculos
```

```python
img = '<nome_da_imagem>.<extenção_da_imagem>'
img = cv.imread(img)# toda a matriz da imagem está na variável, imread lê a imagem para começar o trabalho
```

```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV) # transforma a paleta de cores em HSV
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB) # transforma a paleta de cores em LAB
imgPreBra = cv.cvtColor(img, cv.COLOR_BGR2GRAY) # transforma a paleta de cores em Preto e Branco
```

```python
#Aplicando equalização
img[:,:,0] = cv.equalizeHist(img[:,:,0])# lembra que a imagem é uma matriz, eu estou pasando todas as linhas e todas as colunas e nisso vai me dar um par ordenado Red(0), green(1), blue(2)
img[:,:,1] = cv.equalizeHist(img[:,:,1])
img[:,:,2] = cv.equalizeHist(img[:,:,2])

clahe = cv.createCLAHE(clipLimit = 3.0, tileGridSize = (10,10))# Outro tipo de equalização
img[:,:,0] = clahe.apply(img[:,:,0])# lembra que a imagem é uma matriz, eu estou pasando todas as linhas e todas as colunas e nisso vai me dar um par ordenado Red(0), green(1), blue(2)
img[:,:,1] = clahe.apply(img[:,:,1])
img[:,:,2] = clahe.apply(img[:,:,2])
```

```python
cv.imshow("nome da aba", <nome_da_variável_que_foi_aberto_pelo_OpenCV>)#mostra a imagem
cv.waitKey(0)
cv.destroyAllWindows()# Limpa o bufer do sistema
```

```python
#passa um filtro simples na imagem mas tem outros modos mais eficaz de fazer o mesmo, uso nunpy por conta da otimização dos calculos
np.place(img[:,:,0], img[:,:,0] > 200, [200])
np.place(img[:,:,1], img[:,:,1] > 200, [200])
np.place(img[:,:,2], img[:,:,2] > 200, [200])

np.place(img[:,:,0], img[:,:,0] < 10, [10])
np.place(img[:,:,1], img[:,:,1] < 10, [10])
np.place(img[:,:,2], img[:,:,2] < 10, [10])
```
